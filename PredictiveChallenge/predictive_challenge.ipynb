```python
import matplotlib.pyplot as plt

sex_counts = df['Sex'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sex Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the distribution of each race category
race_counts = df['RaceEthnicityCategory'].value_counts().reset_index()
race_counts.columns = ['RaceEthnicityCategory', 'Count']

# Option 1: Rotate X-Axis Labels
plt.figure(figsize=(10, 6))
sns.barplot(x='RaceEthnicityCategory', y='Count', data=race_counts)
plt.xticks(rotation=45, ha="right")  # Rotate labels for readability
plt.title("Race/Ethnicity Distribution")
plt.show()
```

```python
sns.countplot(data=df, x='Sex', hue='HadHeartAttack')
plt.title("Heart Attack Distribution by Sex")
plt.show()
```

```python
sns.boxplot(data=df, x='HadHeartAttack', y='BMI')
plt.title("BMI Distribution by Heart Attack Status")
plt.show()
```
