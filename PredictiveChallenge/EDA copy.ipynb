Now we want to work on categorical columns so that they can be better used for visualizations. An example would be the age column since it is a string and can therfore not ordered directly. This counts for many other columns as well. So we print each non numerical column to see the values and we will then step by step get rid of them. 

```python
non_numeric_columns = df.select_dtypes(exclude=['number']).columns

# Return values of non numerical columns 
def printUniqueValues(df,column):
    return f'Column: {column} Values: {df[column].unique()}'
def categorical_into_numerocal_yes_no(df,column):
    return  df[column].apply(lambda val: 0 if( val == 'No') else 1 if (val == 'Yes') else  val)
for column in non_numeric_columns:
    row = printUniqueValues(df, column)
    print(row)
```

Now each column will get numerical values for the different categories.

```python
df.loc[:,'Sex_n'] = df['Sex'].apply(lambda val: 0 if( val == 'Male') else 1 ) # 0 if male 1 if female
df.loc[:,'LastCheckupTime_n'] = df['LastCheckupTime'].apply( lambda val: 0 if (val == 'Within past year (anytime less than 12 months ago)')
                                                            else 1 if (val == 'Within past 2 years (1 year but less than 2 years ago)') 
                                                            else 2 if (val == 'Within past 5 years (2 years but less than 5 years ago)')
                                                            else 3 if (val ==  '5 or more years ago')
                                                            else val)
df.loc[:,'RemovedTeeth_n'] = df['RemovedTeeth'].apply(lambda val: 0 if (val == 'None of them') 
                                                        else 1 if (val == '1 to 5')
                                                        else 2 if (val == '6 or more, but not all' )
                                                        else 3 if (val == 'All')
                                                        else val)
df.loc[:,'GeneralHealth_n'] = df['GeneralHealth'].apply(lambda val: 0 if val == 'Poor'
                                                          else 1 if val == 'Fair'  
                                                          else 2 if val == 'Good'
                                                          else 3 if val == 'Very good'
                                                          else 4 if val == 'Excellent'
                                                          else val) 
df.loc[:,'CovidPos_n'] = df['CovidPos'].apply(lambda val: 0 if (val == 'No')
                                                else 1 if (val == 'Tested positive using home test without a health professional')
                                                else 2 if (val == 'Yes')
                                                else val)
df.loc[:,'TetanusLast10Tdap_n'] = df['TetanusLast10Tdap'].apply(lambda val: 0 if (val == 'Yes, received Tdap') 
                                                                else 1 if (val == 'Yes, received tetanus shot, but not Tdap' )
                                                                else 2 if (val == 'Yes, received tetanus shot but not sure what type')
                                                                else 3 if (val == 'No, did not receive any tetanus shot in the past 10 years')
                                                                else val)
df.loc[:,'AgeCategory_n'] = df['AgeCategory'].apply(lambda val: 18 if val == 'Age 18 to 24'
                                                          else 25 if val == 'Age 25 to 29'  
                                                          else 30 if val == 'Age 30 to 34'
                                                          else 35 if val == 'Age 35 to 39'
                                                          else 40 if val == 'Age 40 to 44'
                                                          else 45 if val == 'Age 45 to 49'
                                                          else 50 if val == 'Age 50 to 54'
                                                          else 55 if val == 'Age 55 to 59'
                                                          else 60 if val == 'Age 60 to 64'
                                                          else 65 if val == 'Age 65 to 69'
                                                          else 70 if val == 'Age 70 to 74'
                                                          else 75 if val == 'Age 75 to 79'
                                                          else 80 if val == 'Age 80 or older'
                                                          else val) # Age into numerical category based on start age of range so 18 to 24 will be 18
df.loc[:,'HadDiabetes_n'] = df['HadDiabetes'].apply(lambda val: 0 if val == 'No'
                                                          else 1 if val == 'No, pre-diabetes or borderline diabetes'  
                                                          else 2 if val == 'Yes, but only during pregnancy (female)'
                                                          else 3 if val == 'Yes'
                                                          else val) # scale from 0 to 3 based on value
df.loc[:,'SmokerStatus_n'] = df['SmokerStatus'].apply(lambda val: 0 if val == 'Never smoked'
                                                          else 1 if val == 'Former smoker' 
                                                          else 2 if val == 'Current smoker - now smokes some days'
                                                          else 3 if val == 'Current smoker - now smokes every day'
                                                          else val) # scale from 0 to 4 based on value
df.loc[:,'ECigaretteUsage_n'] = df['ECigaretteUsage'].apply(lambda val: 0 if val == 'Never used e-cigarettes in my entire life'
                                                          else 1 if val == 'Not at all (right now)' 
                                                          else 2 if val == 'Use them some days'
                                                          else 3 if val == 'Use them every day'
                                                          else val) # scale from 0 to 4 based on value

for column in non_numeric_columns:
    unique_values = set(df[column].dropna().unique())
    if unique_values == {'Yes', 'No'} and df[column].isna().any():
       df.loc[:,f'{column}_n'] = categorical_into_numerocal_yes_no(df,column) 
        
```

The next visualization shows the age and gender on the x-axis and the number of heart attacks and anginas. 

```python
# Grouping heart attack data by age category and gender and calculating the sums
heart_attack_counts = df.groupby(['AgeCategory_n', 'Sex_n'])['HadHeartAttack_n'].sum().unstack()

# Create the plot
age_categories = heart_attack_counts.index
bar_width = 0.35  # Width of the bars

# Positions of the bars for the two genders
index = np.arange(len(age_categories))
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)

# Bars for male and female heart attack cases per age group
plt.bar(index, heart_attack_counts[0], bar_width, label='Female')
plt.bar(index + bar_width, heart_attack_counts[1], bar_width, label='Male')

# Axis labels and title
plt.xlabel('Age Category')
plt.ylabel('Count: Had Heart Attack')
plt.title('Number of Heart Attacks by Age Category and Gender')
plt.xticks(index + bar_width / 2, age_categories)  # Center alignment for age category labels
plt.legend()

plt.subplot(1,2,2)

# Grouping angina data by age category and gender and calculating the sums
angina_counts = df.groupby(['AgeCategory_n', 'Sex_n'])['HadAngina_n'].sum().unstack()

# Create the plot
age_categories2 = angina_counts.index
bar_width = 0.35  # Width of the bars

# Positions of the bars for the two genders
index2 = np.arange(len(age_categories2))

# Bars for male and female angina cases per age group
plt.bar(index2, angina_counts[0], bar_width, label='Female')
plt.bar(index2 + bar_width, angina_counts[1], bar_width, label='Male')

# Axis labels and title
plt.xlabel('Age Category')
plt.ylabel('Count: Had Angina')
plt.title('Number of Angina Cases by Age Category and Gender')
plt.xticks(index2 + bar_width / 2, age_categories2)  # Center alignment for age category labels
plt.legend()

plt.show()
```

On the left on right hand side we can clearly see that the number of heart attacks and the number of anginas goes significantly up for older people. This counts for both genders however a little bit more for females. So the age as well as the angina coloum seem to be indicators for heart attacks as well. 

Another interesting column to look at could be the number of heart attacks per smoker category.

```python
# Group heart attack data by smoker status
count_heart_attack_per_smoker_group = df.groupby('SmokerStatus_n')['HadHeartAttack_n'].sum()

# Count values for "no smoker" and "have smoked" only for those who had a heart attack
value_no_smoker_ha = ((df['SmokerStatus_n'] == 0) & (df['HadHeartAttack_n'] == 1)).sum()
value_have_smoked_ha = ((df['SmokerStatus_n'] >= 1) & (df['HadHeartAttack_n'] == 1)).sum()

value_no_smoker = (df['SmokerStatus_n'] == 0).sum()
value_have_smoked = (df['SmokerStatus_n'] >= 1).sum()

# Create the figure and subplots
plt.figure(figsize=(18, 6))

# First plot: Heart attacks per smoker group
plt.subplot(1, 3, 1)
plt.bar(['No Smoker', 'Have Smoked'], [value_no_smoker, value_have_smoked], color='blue')
plt.xlabel('Smoking Category')
plt.ylabel('Count')
plt.title('Number of smokers and non smoker in the entire dataset')

plt.subplot(1, 3, 2)
plt.bar(count_heart_attack_per_smoker_group.index, count_heart_attack_per_smoker_group.values, color='blue')
plt.xlabel('Smoker Status')
plt.ylabel('Frequency of Heart Attacks')
plt.title('Heart Attacks by Smoker Status')
plt.xticks(rotation=45)

# Second plot: Counts of "no smoker" and "have smoked"
plt.subplot(1, 3, 3)
plt.bar(['No Smoker', 'Have Smoked'], [value_no_smoker_ha, value_have_smoked_ha], color='blue')
plt.xlabel('Smoking Category')
plt.ylabel('Count')
plt.title('Heart Attacks by smokers and non smokers')

# Show the plots
plt.tight_layout()
plt.show()
```

On the left hand side we can see that there are much more non smokers in the dataset. So this is not really balanced. 

The figure in the middle is not really as expected. There are a lot of cases of heartattacks for people who dont smoke (value 0). Every other bar represent people who have smoked (value 1), smoke irregular (value 2) or smoke everyday (value 3). We expected to have more cases for value 2 and 3. However if you would make a harder line between 0 and 1-3, it would show that there are more heart attacks for people who have smoked at some point. This can be seen on the right hand side. 

It seems that there is a reltionship between smokers and heart attacks. However we expected a stronger one.

Now we take a look at the BMI. The BMI is converted into 4 groups underweight, Normal weight, overweight and obseity. The number of heart attacks per category will be shown.

```python
df['BMI_categorised'] = df['BMI'].apply(lambda val: 'Underweight' if val <= 18.5 
                                                    else 'Normal weight' if val <= 24.9
                                                    else 'Overweight' if val <= 29.9 
                                                    else 'Obesity' if val >= 30
                                                    else 'Unknown')

bmi_order = ['Underweight', 'Normal weight', 'Overweight', 'Obesity']

df['BMI_categorised'] = pd.Categorical(df['BMI_categorised'], categories=bmi_order, ordered=True)

heart_attacks_per_BMI_category = df.groupby('BMI_categorised',observed = True)['HadHeartAttack_n'].sum()

plt.bar(heart_attacks_per_BMI_category.index, heart_attacks_per_BMI_category.values)
plt.xlabel('BMI Category')
plt.ylabel('Frequency of Heart Attacks')
plt.title('Heart Attacks per BMI Category')
plt.xticks(rotation=45)
plt.show()
```

The figure above clearly shows that the number of heart attacks increases for people with a higher BMI. Especially for the category overweight and obseity.
