```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the dataset (replace 'heart_disease.csv' with your file path
df = pd.read_csv('archive/2020/heart_2020_cleaned.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x='HeartDisease', data=df)
plt.title('Distribution of Heart Disease Presence')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(1,2,1)
sns.histplot(df['AgeCategory'], kde=True)
plt.title('Age Distribution')
plt.xticks(rotation=45)

plt.subplot(1,2,2)
sns.countplot(x='Sex', hue='HeartDisease', data=df)
plt.title('Heart Disease by Gender')
plt.show()
```

```python
plt.subplot(1,4,2)
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Heatmap')

plt.subplot(1,3,2)
sns.boxplot(x='HeartDisease', y='cholesterol', data=df)
plt.title('Cholesterol Levels by Heart Disease Presence')
```
