```python
df = df.drop_duplicates()
df
```

```python
# Lets reduce the size a little bit
selected_columns = [
    'Sex', 'GeneralHealth','PhysicalHealthDays','MentalHealthDays','PhysicalActivities', 'AgeCategory',
    'SleepHours', 'HadHeartAttack', 'HadAngina', 'HadStroke',
    'HadAsthma', 'HadSkinCancer', 'HadCOPD', 'HadDepressiveDisorder', 
    'HadKidneyDisease', 'HadArthritis', 'HadDiabetes', 'SmokerStatus', 
    'ECigaretteUsage', 'ChestScan', 'HeightInMeters', 'WeightInKilograms', 
    'BMI', 'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12', 'PneumoVaxEver'  
]
#Excluded columns: 'LastCheckupTime' 'RemovedTeeth' 'TetanusLast10Tdap', 'CovidPos'
df_small = df[selected_columns].copy()
df_small.head()
```

```python
non_numeric_columns = df_small.select_dtypes(exclude=['number']).columns


# Return values of non numerical columns 
def printUniqueValues(df,column):
    return f'Column: {column} Values: {df[column].unique()}'

for column in non_numeric_columns:
    row = printUniqueValues(df_small, column)
    print(row)
```

```python
# Categorical columns into numerical values
def categorical_into_numerocal_yes_no(df,column):
    return  df[column].apply(lambda val: 0 if( val == 'No') else 1 if (val == 'Yes') else  val)

df_small.loc[:,'Sex_n'] = df_small['Sex'].apply(lambda val: 0 if( val == 'Male') else 1 ) # 0 if male 1 if female
df_small.loc[:,'GeneralHealth_n'] = df_small['GeneralHealth'].apply(lambda val: 0 if val == 'Poor'
                                                          else 1 if val == 'Fair'  
                                                          else 2 if val == 'Good'
                                                          else 3 if val == 'Very good'
                                                          else 4 if val == 'Excellent'
                                                          else val) # scale from 0 to 4 based on value
df_small.loc[:,'AgeCategory_n'] = df_small['AgeCategory'].apply(lambda val: 18 if val == 'Age 18 to 24'
                                                          else 25 if val == 'Age 25 to 29'  
                                                          else 30 if val == 'Age 30 to 34'
                                                          else 35 if val == 'Age 35 to 39'
                                                          else 40 if val == 'Age 40 to 44'
                                                          else 45 if val == 'Age 45 to 49'
                                                          else 50 if val == 'Age 50 to 54'
                                                          else 55 if val == 'Age 55 to 59'
                                                          else 60 if val == 'Age 60 to 64'
                                                          else 65 if val == 'Age 65 to 69'
                                                          else 70 if val == 'Age 70 to 74'
                                                          else 75 if val == 'Age 75 to 79'
                                                          else 80 if val == 'Age 80 or older'
                                                          else val) # Age into numerical category based on start age of range so 18 to 24 will be 18
df_small.loc[:,'HadDiabetes_n'] = df_small['HadDiabetes'].apply(lambda val: 0 if val == 'No'
                                                          else 1 if val == 'No, pre-diabetes or borderline diabetes'  
                                                          else 2 if val == 'Yes, but only during pregnancy (female)'
                                                          else 3 if val == 'Yes'
                                                          else val) # scale from 0 to 3 based on value
df_small.loc[:,'SmokerStatus_n'] = df_small['SmokerStatus'].apply(lambda val: 0 if val == 'Never smoked'
                                                          else 1 if val == 'Former smoker' 
                                                          else 2 if val == 'Current smoker - now smokes some days'
                                                          else 3 if val == 'Current smoker - now smokes every day'
                                                          else val) # scale from 0 to 4 based on value
df_small.loc[:,'ECigaretteUsage_n'] = df_small['ECigaretteUsage'].apply(lambda val: 0 if val == 'Never used e-cigarettes in my entire life'
                                                          else 1 if val == 'Not at all (right now)' 
                                                          else 2 if val == 'Use them some days'
                                                          else 3 if val == 'Use them every day'
                                                          else val) # scale from 0 to 4 based on value
```

```python
for column in non_numeric_columns:
    unique_values = set(df_small[column].dropna().unique())
    if unique_values == {'Yes', 'No'} and df_small[column].isna().any():
       df_small.loc[:,f'{column}_n'] = categorical_into_numerocal_yes_no(df_small,column) 
df_small
```

```python
only_numeric_df = df_small.select_dtypes(include='number')
correlation_matrix = only_numeric_df.corr()

plt.figure(figsize=(20, 12))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```

```python
sex_counts = df_small['Sex'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Sex Distribution')
plt.show()
```

```python
heart_attack_counts_per_gender = df_small.groupby('Sex_n')['HadHeartAttack_n'].sum()
count_heart_attakcks = df_small['HadHeartAttack'].value_counts()

plt.figure(figsize=(12,8))
plt.subplot(1,2,1)
plt.bar(count_heart_attakcks.index,count_heart_attakcks.values,color=['orange','blue'])
plt.xlabel('Had heart attack')
plt.ylabel('Count')
plt.title('Number of heart attacks')

plt.subplot(1,2,2)
plt.bar(df_small['Sex'].unique(),heart_attack_counts_per_gender, color=['pink','blue'])
plt.xlabel('Gender')
plt.ylabel('Count of heart attacks')
plt.title('Distribution per gender of heart attakcks')

plt.show()
```

```python
# Gruppieren der Herzinfarktdaten nach Alterskategorie und Geschlecht und Berechnen der Summen
heart_attack_counts = df_small.groupby(['AgeCategory_n', 'Sex_n'])['HadHeartAttack_n'].sum().unstack()

# Plot erstellen
age_categories = heart_attack_counts.index
bar_width = 0.35  # Breite der Balken

# Positionen der Balken für die beiden Geschlechter
index = np.arange(len(age_categories))
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)

# Balken für männliche und weibliche Herzinfarktfälle pro Altersgruppe
plt.bar(index, heart_attack_counts[0], bar_width, label='Female')
plt.bar(index + bar_width, heart_attack_counts[1], bar_width, label='Male')

# Achsenbeschriftungen und Titel
plt.xlabel('Age Category')
plt.ylabel('Count: Had heart attack')
plt.title('Number of Heart Attacks by Age Category and Gender')
plt.xticks(index + bar_width / 2, age_categories)  # Mittige Positionierung der Alterskategorie-Labels
plt.legend()
#plt.show()
plt.subplot(1,2,2)
# Gruppieren der Herzinfarktdaten nach Alterskategorie und Geschlecht und Berechnen der Summen
angina_counts = df_small.groupby(['AgeCategory_n', 'Sex_n'])['HadAngina_n'].sum().unstack()
# Plot erstellen
age_categories2 = angina_counts.index
bar_width = 0.35  # Breite der Balken
# Positionen der Balken für die beiden Geschlechter
index2 = np.arange(len(age_categories2))
# Balken für männliche und weibliche Herzinfarktfälle pro Altersgruppe
plt.bar(index2, angina_counts[0], bar_width, label='Female')
plt.bar(index2 + bar_width, angina_counts[1], bar_width, label='Male')

# Achsenbeschriftungen und Titel
plt.xlabel('Age Category')
plt.ylabel('Count: Had angina')
plt.title('Number of angina by Age Category and Gender')
plt.xticks(index2 + bar_width / 2, age_categories2)  # Mittige Positionierung der Alterskategorie-Labels
plt.legend()

plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(15, 4), sharey=True)
# Plot for BMI
sns.histplot(df_small['BMI'], bins=30, kde=True, ax=axes[0])
axes[0].set_title("Distribution of BMI")
axes[0].set_xlabel("BMI")
axes[0].set_ylabel("Frequency")

# Plot for PhysicalHealth
sns.histplot(df_small['PhysicalHealthDays'], bins=30, kde=True, ax=axes[1])
axes[1].set_title("Distribution of Physical Health")
axes[1].set_xlabel("Physical Health Days")

# Plot for MentalHealth
sns.histplot(df_small['MentalHealthDays'], bins=30, kde=True, ax=axes[2])
axes[2].set_title("Distribution of Mental Health")
axes[2].set_xlabel("Mental Health Days")

# Plot for SleepTime
sns.histplot(df_small['SleepHours'], bins=30, kde=True, ax=axes[3])
axes[3].set_title("Distribution of Sleep Time")
axes[3].set_xlabel("Sleep Time (hours)")

plt.tight_layout()
plt.show()
```

```python
#heart attack per smocking group
count_heart_attack_per_smoker_group = df_small.groupby('SmokerStatus_n')['HadHeartAttack_n'].sum()

plt.bar(count_heart_attack_per_smoker_group.index,count_heart_attack_per_smoker_group.values)
plt.xlabel('Smoker Status')
plt.ylabel('Frequenz Heart attacks')
plt.xticks(rotation=45)
plt.show()
```

```python
df_small['BMI_categorised'] = df_small['BMI'].apply(lambda val: 'Underweight' if val <= 18.5 
                                                    else 'Normal weight' if val <= 24.9
                                                    else 'Overweight' if val <= 29.9 
                                                    else 'Obesity' if val >= 30
                                                    else 'Unknown')

bmi_order = ['Underweight', 'Normal weight', 'Overweight', 'Obesity']

df_small['BMI_categorised'] = pd.Categorical(df_small['BMI_categorised'], categories=bmi_order, ordered=True)

heart_attacks_per_BMI_category = df_small.groupby('BMI_categorised',observed = True)['HadHeartAttack_n'].sum()

plt.bar(heart_attacks_per_BMI_category.index, heart_attacks_per_BMI_category.values)
plt.xlabel('BMI Category')
plt.ylabel('Frequency of Heart Attacks')
plt.title('Heart Attacks per BMI Category')
plt.xticks(rotation=45)
plt.show()
```
